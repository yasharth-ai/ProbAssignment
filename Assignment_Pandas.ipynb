```python
import pandas as pd
import numpy as np
from scipy import stats
from numpy import percentile
df = pd.read_csv('DataSets\\titanic.csv')
print("Max Age-   " + str(max(df['Age'].dropna())))
print("Min Age-   " + str(min(df['Age'].dropna())))
print("Numpy Median-   " + str(np.median(df['Age'].dropna())))
# calculate interquartile range
q25, q75 = percentile(df['Age'].dropna(), 25), percentile(df['Age'].dropna(), 75)
print("Q25-   " + str(q25))
print("Q75-   " + str(q75))
```

```python
import pandas as pd
df = pd.read_csv('DataSets\\titanic.csv')

#Relapcing string Columns with 0,1,2 values like Sex, Embarked
def mapping(data,feature):
    featureMap=dict()
    count=0
    for i in sorted(data[feature].unique(), reverse=True):
        featureMap[str(i)]=count
        count=count+1
    data[feature]=data[feature].map(featureMap)
    return data

df = mapping(df,"Sex")
df["Embarked"] = df["Embarked"].fillna('0')
df = mapping(df,"Embarked")
df["Cabin"] = df["Cabin"].fillna('0')
df = mapping(df,"Cabin")

#Filling default values with 0
df["Fare"] = df["Fare"].fillna(0)
df["Age"] = df["Age"].fillna(0)

df=df.drop(["PassengerId","Name","Ticket"],axis=1)

#Standard Deviation Normalization
normalized_df_std=(df-df.mean())/df.std()
normalized_df_std

#Max Min Normalization on scale of 10
normalized_df=(df-df.min())/(df.max()-df.min())*10
normalized_df
```

```python
import pandas as pd
import datetime
ser = pd.Series(['01 Jan 2010','02-02-2010','20120303','2013/04/04','2014-05-05','2015-06-06T12:20'])
dl = pd.DataFrame(pd.to_datetime(ser.tolist()), columns=['Date'])
dl['Year']=  dl['Date'].dt.year
dl['Month']=  dl['Date'].dt.month
dl['Day']=  dl['Date'].dt.day
dl['Week']=  dl['Date'].dt.week
dl['DayofYear'] = dl['Date'].dt.dayofyear 
dl['DayofWeek'] = dl['Date'].dt.dayofweek
dl
```

```python
import pandas as pd
import datetime
from datetime import datetime
from datetime import timedelta
startvalue =  "2000-01-01"
startDate = datetime.strptime(startvalue, '%Y-%m-%d')
dl_Weekend =[]
weekendcounter = 0
while (weekendcounter < 10):
    if(startDate.weekday() > 4):        
        dl_Weekend.append(startDate)
        weekendcounter = weekendcounter + 1
    startDate = startDate + timedelta(days=1)    
dl_Weekend
```
